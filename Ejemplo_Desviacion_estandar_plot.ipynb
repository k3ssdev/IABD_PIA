```python
import matplotlib.pyplot as plt
import numpy as np

# Datos del ejemplo
valores = [65, 70, 75, 80, 85]
media = 75
desviaciones = [valor - media for valor in valores]

# Crear el gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Gráfico de los valores
ax.plot(valores, 'o-', label='Valores')
ax.axhline(y=media, color='r', linestyle='--', label='Media (75)')

# Dibujar las desviaciones con líneas
for i, (valor, desviacion) in enumerate(zip(valores, desviaciones)):
    ax.plot([i, i], [media, valor], color="gray", linestyle="--")
    ax.text(i, (valor + media) / 2, f"{desviacion:+}", ha="center", va="center", color="gray")

# Configuraciones adicionales
ax.set_xticks(range(len(valores)))
ax.set_xticklabels([f"Estudiante {i+1}" for i in range(len(valores))])
ax.set_ylabel("Calificación")
ax.set_title("Calificaciones y Desviaciones respecto a la Media")
ax.legend()
plt.grid(axis='y')

plt.show()
```

```python
# prompt: Calificaciones y Desviaciones respecto a la Media con seaborn

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Datos del ejemplo
valores = [65, 70, 75, 80, 85]
media = np.mean(valores)  # Calcular la media directamente
desviaciones = [valor - media for valor in valores]

# Crear el gráfico usando Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=range(len(valores)), y=valores, label='Valores', marker='o', s=100)
plt.axhline(y=media, color='r', linestyle='--', label=f'Media ({media:.2f})')
plt.plot(range(len(valores)), valores, color='blue', linestyle='-', label='Conexión de Estudiantes')

# Dibujar las desviaciones con líneas
for i, (valor, desviacion) in enumerate(zip(valores, desviaciones)):
    plt.plot([i, i], [media, valor], color="gray", linestyle="--")
    plt.text(i, (valor + media) / 2, f"{desviacion:.2f}", ha="center", va="center", color="gray")

# Configuraciones adicionales
plt.xticks(range(len(valores)), [f"Estudiante {i+1}" for i in range(len(valores))])
plt.ylabel("Calificación")
plt.xlabel("Estudiante")
plt.title("Calificaciones y Desviaciones respecto a la Media")
plt.legend()
plt.grid(axis='y')

plt.show()
```
